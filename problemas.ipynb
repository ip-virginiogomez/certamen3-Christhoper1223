```python
ventas = [
    [70000, 25000, 32000],
    [55000, 67000, 44000],
    [23000, 99000, 17000]
]
totales = []
for v in range(3):
    suma_vendedor = sum(ventas[v])
    totales.append(suma_vendedor)
    print(f"Total Vendedor {v + 1}: ${suma_vendedor}")

mejor = 0
for v in range(1, 3):
    if totales[v] > totales[mejor]:
        mejor = v

print(f"El vendedor con mayor rendimiento es el Vendedor {mejor + 1}")

for v in range(3):
    if totales[v] < 30000:
        print(f"Vendedor {v + 1} tuvo bajo desempeño (Menos de $30.000)")
```
